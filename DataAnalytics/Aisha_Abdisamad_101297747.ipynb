**Q3-3 (1 pts)** - How many movies were released for each year between 1960 and 2020? Create a line chart showing the trend of the movie release count over time (do not include data after 2020). The chart should meet the following requirements:
- The figure size should be `(12, 4)`
- x-axis shows time and has a label `Year`
- y-axis shows the release count and has a label `Count`
- Set the figure title as `Movie Release Count in 60 Years: 1960 to 2020`.
- Use circle marker, the line should be solid line, and color of the line should be orange

```python
df['release_year'] = df['release_date']
# use the pandas.to_datetime to convert the release date to a release year. This function converts a scalar, array-like, Series or DataFrame to a pandas datetime object.
df['release_year'] = pd.to_datetime(df['release_year']).dt.year
#set the 60 year range by setting release year greater than or equal to the year 1960 and less than the year 2020
released_movies = df[(df['release_year']>= 1960) & (df['release_year']< 2020)]
num_movies_released = released_movies['release_year'].value_counts().sort_index()

plt.figure(figsize=(12,4))
plt.plot(num_movies_released.index, num_movies_released.values, marker="o", linestyle="-", color="orange")
plt.xlabel("Year")
plt.ylabel("Count")
plt.title("Movie Release Count in 60 Years: 1960 to 2020")
plt.show()
```

**Q3-4 (2 pts)** - This is a follow up question of Q3-4. 

What are the average profits of movies released for each year for each language between 1960 and 2020? Create a line chart showing the trend of the movie average profits over time (Still, do not include data after 2020) for each language. So, in total, you should have 10 lines in the chart. 

The chart should meet the following requirements:
- The figure size should be `(12, 6)`
- x-axis shows time and has a label `Year`
- y-axis shows the average profits and has a label `Average Profits per Movie`
- Set the figure title as `Average Profits per Movie by Language in 60 Years: 1960 to 2020`.
- The line should be solid line (do not apply markers in this chart.)
- Each line should have a different color in the `Tableau Palette` (You can find color list [HERE](https://matplotlib.org/stable/gallery/color/named_colors.html).)
- Show legend for all lines

Check the chart and what do you find? Anything interesting? Any insights? Identify 3 trends/insights/patterns from the chart and write them in a markdown cell following your code cell below.

```python
#create a list of tableau colours
color_list = ['tab:blue','tab:orange','tab:green','tab:red','tab:purple','tab:brown','tab:pink','tab:gray','tab:olive','tab:cyan']
top_ten_languages = df['language'].value_counts().head(10).index

plt.figure(figsize=(12,6))

color_index = 0
#use a for loop to loop through the language dictonary created earlier in Q2-5, this loop will output a new line as it loops through the languages while updating each line with a new color from the color list
for language in top_ten_languages:
    df_language = released_movies[released_movies['language'] == language].groupby('release_year')['profit'].mean().sort_index()
    plt.plot(df_language.index, df_language.values, linestyle="-", color= color_list[color_index],label = language)
    color_index+=1
    
plt.xlabel("Year")
plt.ylabel("Average Profits per Movie")
plt.title("Average Profits per Movie by Language in 60 Years: 1960 to 2020")   
plt.legend()
plt.grid(True)
plt.show()
```

Three trend/insights/patterns I have noticed from the chart are:

- The blue line that represents English movies shows how English movies are steadily dominating as they have the highest profits over time. Espically after the 1970's where they continue to steadily grow.
- Non-english movies see dramatic increases and spikes in profits after and during the year 2000. Chinese movies see a large spike in the mid 2010s. This is likely because streaming services giants like Netflix started streaming Chinese movies in the mid 2010s resulting in an increased number of views of these movies.
- If we look at the red line representing Japanese movies we see an initial spike in the late 1990 and an even bigger spike in the early 2000. I belive this is because anime which is a from of Japanese animation gain popularity globably around that time leading to increased profits in that sector.
